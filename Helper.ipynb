```python
# Extract labeled data
data_frames = []
file_name = '202312.parquet'
data_frames.append(pd.read_parquet(file_name))

data = pd.concat(data_frames, ignore_index=True)


irrelevant_categories = ['020: Street Lighting: HOEP', '018: Unmetered Scattered Load', '021: Street Lighting: Tiered']
data = data[~data['RATECLASS_DESC'].isin(irrelevant_categories)]
data['# of Chargers'] = pd.to_numeric(data['# of Chargers'], errors='coerce')


labeled_data = data.dropna(subset=['# of Chargers'])
unlabeled_data = data[data['# of Chargers'].isna()].drop(columns=['# of Chargers'])


output_file_labeled = 'labeled_data_12.parquet'
labeled_data.to_parquet(output_file_labeled, engine='pyarrow')
```

```python
#combine data
combined_data_frames = []
for month in range(1, 13):
    file_name = f'2023{month:02d}.parquet'
    combined_data_frames.append(pd.read_parquet(file_name))

combined_data = pd.concat(combined_data_frames, ignore_index=True)

# Step 2: Save the combined DataFrame to a new Parquet file
output_file_combined = 'combined_data.parquet'
combined_data.to_parquet(output_file_combined, engine='pyarrow')
```

```python
# xlsx to prequet
file_path = '202309.xlsx'
xls = pd.ExcelFile(file_path)
sheet_names = xls.sheet_names
data_df = pd.read_excel(file_path, sheet_name=sheet_names[0])
column_names = data_df.columns.tolist()
dfs = [data_df]
for sheet_name in sheet_names[1:]:  # Skip the first sheet since it's already read
    df = pd.read_excel(file_path, sheet_name=sheet_name, header=None, names=column_names)
    dfs.append(df)
combined_df = pd.concat(dfs, ignore_index=True)
combined_df.to_parquet('202309.parquet')
```
